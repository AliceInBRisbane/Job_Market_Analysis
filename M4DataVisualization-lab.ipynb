## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# Import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

QUERY = """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY,conn)


# Plot the histogram for the 'ConvertedComp' column
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=10, edgecolor='black', color='lightblue')
plt.title('Histogram of Converted Annual USD Salaries')
plt.xlabel('Converted Annual USD Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

plt.figure(figsize=(10, 6))  # Set the figure size
plt.boxplot(df['Age'].dropna(), vert=True)  # Drop null values before plotting
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.ylabel('Year')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here


# Create the scatter plot
plt.scatter(df['Age'], df['WorkWeekHrs'], s=10,color='darkblue')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here


# Create the bubble plot
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*5, alpha=0.5)  # Adjust the size multiplier as needed
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs with Age as Bubble Size')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here


QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df_database = pd.read_sql_query(QUERY,conn)

# Get the top 5 databases
top_5_databases = df_database['DatabaseDesireNextYear'].value_counts().head(5).index
top_5_counts = df_database['DatabaseDesireNextYear'].value_counts().head(5).values

# Create the pie chart
plt.pie(top_5_counts, labels=top_5_databases, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_wk = pd.read_sql_query(QUERY,conn)

df_g=df_wk.groupby('Age').median()

df_g.plot(kind='bar',stacked=True)
plt.title('Median Workweek Hrs and CodeRevHrs in 30-35 Age Group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age,ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_cc = pd.read_sql_query(QUERY,conn)

df_ccm=df_cc.groupby('Age').median()

df_ccm.plot(kind='line')
plt.title('Median salary in 45-60 age group')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df_mb = pd.read_sql_query(QUERY,conn)
df_mbh=df_mb['MainBranch'].value_counts().rename_axis('MainBranch').reset_index(name='Counts')

df_mbh.plot(kind='barh')
plt.title('MainBranch')
plt.show()
```
